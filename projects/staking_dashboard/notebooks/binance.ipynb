```python
import requests
import json
import pandas as pd
from datetime import datetime
```

```python
url = "https://api.binance.com/api/v1/klines?symbol=ONEUSDT&interval=1d"
r = requests.get(url)
content = json.loads(r.content)
```

```python
df = pd.DataFrame(content, columns = ['Open time', 'Open', 'High', 'Low', 'Close', 'Volume', 'Close time', 'Quate asset volume', 'Number of trades', 'Taker buy base asset volume', 'Taker buy quote asset volume', 'Ignore'])
```

```python
df['Open time'] = df['Open time'].apply(lambda c: datetime.fromtimestamp(c/1000).strftime('%Y_%m_%d %H:%M:%S'))
df['Close time'] = df['Close time'].apply(lambda c: datetime.fromtimestamp(c/1000).strftime('%Y_%m_%d %H:%M:%S'))
```

```python
df.to_csv('../csv/binance.csv')
```
