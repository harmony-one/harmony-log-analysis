```python
import json 
import pandas as pd
import numpy as np
import re
from datetime import datetime
import os
from os import path
from collections import defaultdict
```

```python
followers = defaultdict(int)
count = defaultdict(int)
link = defaultdict(str)
url = defaultdict(str)
log_dir = "friends/"
files = os.listdir(log_dir)

for file in files:
    with open(log_dir + file, errors='ignore') as f:
        for line in f.readlines():
            try: 
                data = json.loads(line)
                name = data['screen_name']
                followers[name] += 1
                if name not in count:
                    follower_count = data['followers_count']               
                    count[name] = follower_count
                    link[name] = "https://twitter.com/{}".format(name)
                    url[name] = data['url']
                    description[name] = data['description']
            except:  
                pass
```

```python
df_0 = pd.DataFrame(followers.items(), columns=['name', '#followed_by_harmony'])
df_1 = pd.DataFrame(count.items(), columns=['name', '#followers'])
df_2 = pd.DataFrame(link.items(), columns=['name', 'twitter_link'])
df_3 = pd.DataFrame(url.items(), columns=['name', 'harmony_url'])
df_4 = pd.DataFrame(url.items(), columns=['name', 'description'])

df = df_0.join(df_1.set_index('name'), on = 'name')
df = df.join(df_2.set_index('name'), on = 'name')
df = df.join(df_3.set_index('name'), on = 'name')
df = df.join(df_4.set_index('name'), on = 'name')
```

## filter those with harmony url

```python
pattern = re.compile('.*?\/#.*?=one.*?')
df['hmy-url-flag'] = df['harmony_url'].apply(lambda c: True if c and re.match(pattern, c) else False)
df['description_flag'] = df['description'].apply(lambda c: True if c and re.match(pattern, c) else False)
```
