```python
import json
import pandas as pd
import numpy as np
import csv
import re
```

```python
df = pd.read_csv('json/angels-20200702', delimiter=r"\s+", header = None, error_bad_lines=False,names = ['#followe_by_team', 'retweet_count','favorite_count','tweet_created_at','followers_count','friends_count','followers_count/friends_count+1','statuses_count','favorites_count','id','account_created_at','screen_name','name', 'url', 'url_2', 'url_3', 'url_4', 'url_5', 'url_6']) 
```

## merge the urls

```python
pattern = re.compile('.*?\/#.*?=one.*?')
column = ['url', 'url_2', 'url_3', 'url_4', 'url_5', 'url_6']
for url in column:
    df[url] = df[url].apply(lambda c: c if type(c) == str and re.match(pattern, c) else np.nan)
    
df['url'].fillna(df['url_2'], inplace = True)
df['url'].fillna(df['url_3'], inplace = True)
df['url'].fillna(df['url_4'], inplace = True)
df['url'].fillna(df['url_5'], inplace = True)
df['url'].fillna(df['url_6'], inplace = True)

df.drop(['url_2', 'url_3', 'url_4', 'url_5', 'url_6'], axis = 1, inplace=True)
```

## parse the wallet

```python
df['wallet'] = df['url'].apply(lambda c: c.split("=")[1] if type(c) == str else np.nan)
```
